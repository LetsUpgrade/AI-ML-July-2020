Tuples
--
> Tuple is similar to a list. However, in tuples, we cannot change the elements after assignment i.e they are immutable.

> Each element or value inside the tuple is called an item.

```python
# type your code here2=6
s = ('y','v','angle')
print(s)
print(type(s))
```

```python
# empty tuple
# type your code here2..8
t = ()
print(t)
print(type(t))
```

```python
# type your code here3
t = (1.0)
print(t)
print(type(t))
```

```python
# Single value tuple must use a __________________
# type your code here

t=(1,)
print(t)
print(type(t))
```

```python
l = [1.0]
print(type(l))
```

```python
# type your code here3+3
t = (1,2,33.4443,5,6)
# type your code here
print(t[1:3])
print(t[1:])
print(t[:4])
print(t[::-1])
print(t[-5:-1])
```

Variable Length Arguments.
--

```python
def var(*args):
    print(args)

# type your code here4%5
var(1)
var(1,2,3)
var("darshan","Lets Upgrade",32)
var()
```
